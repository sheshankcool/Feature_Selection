```python
# Set 'price' as the target variable
y = df['price']

# Extract the input features
X = df.drop(['price'], axis=1)
```

```python
li=[]
for i in X.columns:
    corr=list(stats.pearsonr(x=df[i],y=df["price"]))
    li.append([i,corr[0],corr[1]])
cdf=pd.DataFrame(li,columns=["feature","correlation","p-value"])
cdf.sort_values(by="p-value",ascending=True)
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

selector = SelectKBest(f_regression, k=5)
X_important = selector.fit_transform(X, y)

# Get a boolean mask of the selected features
mask = selector.get_support()

# Create a list of the selected feature names
important_feature_names = X.columns[mask]

print(important_feature_names)
```
